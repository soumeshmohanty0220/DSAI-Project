```python
import pandas as pd
```

```python
df = pd.read_csv("../Dataset/country.csv")
```

```python
df.head()
```

```python
for i in df.columns:
    if i=="Country":
        continue
    else:
        df.rename(columns={i: pd.to_datetime(i)}, inplace=True)
```

```python
df.head()
```

```python
df=df.transpose()
```

```python
df.head()
```

```python
df =df.set_axis(df.iloc[0], axis=1)
df =df[1:]
```

```python
df.head()
```

```python
df.shape
```

```python
df.rename(columns={"Country": "Date"}, inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.ylim(0,103802702)
for i in df.columns:
    lst = df[i].tolist()
    plt.plot(lst, label=i)
plt.xlabel("Days")
plt.ylabel("No. of cases")
plt.legend()
plt.show()
```

```python
for i in df.columns:
    plt.figure(figsize=(4,2))
    plt.ylim(0,103802702)
    lst = df[i].tolist()
    plt.plot(lst, label=i)
    plt.xlabel("Days")
    plt.ylabel("No. of cases")
    plt.legend()
    plt.show()
```
